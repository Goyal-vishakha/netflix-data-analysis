```python
# changing the data type od date_added column
data['date_added']=pd.to_datetime(data['date_added'], format='%B %d, %Y')
```

```python
data.head(10)
```

```python
#dealing with the missing values in the director,cast, country,rating by using fillna function by filling "Unavaliable".
```

```python
data.fillna({"director":"Unavaliable","cast":"Unavaliable","country":"Unavaliable","rating":"Unavaliable"},inplace=True)
```

```python
data.head(10)
```

```python
# we will replace the date in date_added from the most recently used date in the date_added.
```

```python
most_recent_date=data["date_added"].max()
```

```python
most_recent_date
```

```python
data.fillna({'date_added':most_recent_date},inplace=True)
```

# data visualization

```python
# to know which type of show are mostly watched on netflix
```

```python
data.type.value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='type',data=data,label=True)
plt.title("Count Vs Type of shows")
plt.show()
```

```python
# to check type of content based on country
```

```python
data.country.value_counts().head(20)
```

```python
plt.figure(figsize=(20,20))
sns.countplot(x='country',order=data.country.value_counts().index[0:20],data=data)
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.countplot(y='country',order=data.country.value_counts().index[0:20],data=data)
plt.show()
```

```python
# we can conclude that united state produce the highest content on the netflix . these are top 20 counnrty 
```
